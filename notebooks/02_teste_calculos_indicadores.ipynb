# 2 - Teste Cálculos Indicadores

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_csv("../dados_tratados/teste_restinga_dados_preparados.csv")

print("Dimensão original df:", df.shape)
```

```python
df.info() 
```

Conversão de datas:

```python
df["Mês De Ocorrência da Situação"] = pd.to_datetime(
    df["Mês De Ocorrência da Situação"],
    dayfirst=True,
)

df["Data de Fim Previsto do Ciclo"] = pd.to_datetime(
    df["Data de Fim Previsto do Ciclo"],
    dayfirst=True,
)

df["Data de Inicio do Ciclo"] = pd.to_datetime(
    df["Data de Inicio do Ciclo"],
    dayfirst=True,
)

df["Data de Ocorrencia da Matricula"] = pd.to_datetime(
    df["Data de Ocorrencia da Matricula"],
    dayfirst=True,
    errors='coerce'
)

df.dtypes
```

Testando alguns agrupamentos:

```python
contagem_matriculas_ano_curso = df.groupby(["Ano", "Nome de Curso"])["Código da Matricula"].count().reset_index() #resetamos o index para permitir agrupar/salvar em excel/csv
print(contagem_matriculas_ano_curso)
```
