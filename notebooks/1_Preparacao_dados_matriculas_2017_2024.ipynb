 ## 1 - Preparação dos Dados
Fonte: Microdados de Matrículas da Plataforma Nilo Peçanha (2017-2024)

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

#### 1.1 Transformação dos arquivos .csv em Data Frames, selecionando as colunas a serem utilizadas:

```python
selecao = [
    "Ano", "Carga Horaria", "Carga Horaria Mínima", "Categoria da Situação", 
    "Cor / Raça", "Código da Matricula", "Código do Ciclo Matricula", 
    "Data de Fim Previsto do Ciclo", 
    "Data de Inicio do Ciclo", "Data de Ocorrencia da Matricula", 
    "Eixo Tecnológico", "Faixa Etária", "Fator Esforço Curso", 
    "Idade", "Instituição", 
    "Modalidade de Ensino", "Município", "Mês De Ocorrência da Situação", 
    "Nome de Curso", "Renda Familiar", "Sexo", 
    "Situação de Matrícula", "Subeixo Tecnológico", "Tipo de Curso", 
    "Tipo de Oferta", "Total de Inscritos", "Turno", "UF", "Unidade de Ensino"
]


df_2017 = pd.read_csv("../dados_brutos/microdados_matriculas_2017.csv", sep=";", usecols=selecao)
df_2018 = pd.read_csv("../dados_brutos/microdados_matriculas_2018.csv", sep=";", usecols=selecao)
df_2019 = pd.read_csv("../dados_brutos/microdados_matriculas_2019.csv", sep=";", usecols=selecao)
df_2020 = pd.read_csv("../dados_brutos/microdados_matriculas_2020.csv", sep=";", usecols=selecao)
df_2021 = pd.read_csv("../dados_brutos/microdados_matriculas_2021.csv", sep=";", usecols=selecao)
df_2022 = pd.read_csv("../dados_brutos/microdados_matriculas_2022.csv", sep=";", usecols=selecao)
df_2023 = pd.read_csv("../dados_brutos/microdados_matriculas_2023.csv", sep=";", usecols=selecao)
df_2024 = pd.read_csv("../dados_brutos/microdados_matriculas_2024.csv", sep=";", usecols=selecao)


print("Dimensão original 2017:", df_2017.shape)
print("Dimensão original 2018:", df_2018.shape)
print("Dimensão original 2019:", df_2019.shape)
print("Dimensão original 2020:", df_2020.shape)
print("Dimensão original 2021:", df_2021.shape)
print("Dimensão original 2022:", df_2022.shape)
print("Dimensão original 2023:", df_2023.shape)
print("Dimensão original 2024:", df_2024.shape)
```

#### 1.2 Concatenação dos Data Frames gerados e exibição dos primeiros e últimos registros:

```python
df_2017_2024 = pd.concat([df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024])

df_2017_2024.head()
```

```python
df_2017_2024.tail()
```

#### 1.3 Aplicação de Filtro de Unidade de Ensino = Campus Porto Alegre Restinga:

```python
df_restinga = df_2017_2024.loc[df_2017_2024["Unidade de Ensino"] == "Campus Porto Alegre Restinga"]
print("Após filtro Campus Restinga:", df_restinga.shape)
```

#### 1.4 Exclusão de colunas desnecessárias:

```python
df_restinga = df_restinga.drop(columns=["Instituição", "Município", "UF", "Unidade de Ensino"])
print("Após exclusão de colunas desnecessárias:", df_restinga.shape)
```

#### 1.5 Verificação dos nomes dos cursos do Campus Restinga:

```python
df_restinga.groupby("Nome de Curso").size()
```

#### 1.6 Seleção dos cursos Técnicos e Superiores do Campus Restinga:

```python
df_restinga["Tipo de Curso"].unique()
```

```python
df_restinga = df_restinga[
    df_restinga["Tipo de Curso"].isin(["Licenciatura", "Técnico", "Tecnologia"])
]
```

```python
df_contagem = df_restinga.groupby("Nome de Curso").size()
df_contagem
```

```python
df_restinga["Código da Matricula"].count()
```

#### 1.7 Limpeza dos cursos extintos e padronização dos nomes:

```python
cursos_excluir = [
    "Técnico em Administração",
    "Técnico em Informática Para Internet",
    "Técnico em Informática para Internet",
    "Técnico em Recursos Humanos",
    "Técnico em Redes de Computadores"
]

df_restinga = df_restinga[
    ~df_restinga["Nome de Curso"].isin(cursos_excluir)
]

df_restinga.loc[:, "Nome de Curso"] = df_restinga["Nome de Curso"].replace({
    "Gestão desportiva e de Lazer": "Gestão Desportiva e de Lazer",
    "Letras": "Letras - Português e Espanhol",
    "Letras - Língua Estrangeira": "Letras - Português e Espanhol"
})

df_restinga["Nome de Curso"].unique()
```

#### 1.9 Verificando os valores ausentes:

```python
df_restinga.isnull().sum()   # df_restinga.isna().sum()  #esse comando é equivalente
```

#### 1.10 Tratamento dos valores ausentes:

```python
#retirar coluna total inscritos, porque não será necessária ao cálculo dos indicadores
df_restinga = df_restinga.drop(columns=["Total de Inscritos"])
print("Após exclusão de total de inscritos:", df_restinga.shape)
```

```python
#retirar as 2 linhas com valores nulos do mês de ocorrência

df_restinga.dropna(subset=["Mês De Ocorrência da Situação"], inplace=True) #inplace=True para salvar a retirada dos nulos --- verificar se faremos assim

print("Após exclusão de linhas desnecessárias:", df_restinga.shape)
```

#### 1.12 Convertendo os tipos de dados:

```python

df_restinga["Data de Fim Previsto do Ciclo"] = pd.to_datetime(
    df_restinga["Data de Fim Previsto do Ciclo"],
    dayfirst=True,
    errors='coerce'
)

df_restinga["Data de Inicio do Ciclo"] = pd.to_datetime(
    df_restinga["Data de Inicio do Ciclo"],
    dayfirst=True,
    errors='coerce'
)
df_restinga["Data de Ocorrencia da Matricula"] = pd.to_datetime(
    df_restinga["Data de Ocorrencia da Matricula"],
    dayfirst=True,
    errors='coerce'
)

df_restinga["Mês De Ocorrência da Situação"] = pd.to_datetime(
    df_restinga["Mês De Ocorrência da Situação"],
    dayfirst=True,
    errors='coerce'
)

df_restinga.dtypes
```

#### Ajustes das situações de matrícula:

```python
df_restinga["Situação de Matrícula"].unique()
```

```python
# ajuste matrículas Em curso para Retido ou Em fluxo:
periodo_fim_previsto = df_restinga["Data de Fim Previsto do Ciclo"].dt.to_period("M")
periodo_referencia = pd.PeriodIndex.from_fields(
    year=df_restinga["Ano"],
    month=12,
    freq="M"
)

df_restinga.loc[df_restinga["Situação de Matrícula"] == "Em curso", "Situação de Matrícula"] = np.where(
    periodo_fim_previsto[df_restinga["Situação de Matrícula"] == "Em curso"] < periodo_referencia[df_restinga["Situação de Matrícula"] == "Em curso"],
    "Retido",
    "Em fluxo"
)

# ajuste matrículas Concluída para Concluída no prazo ou Concluída com atraso:
periodo_conclusao = df_restinga["Mês De Ocorrência da Situação"].dt.to_period("M")

# Diferença em meses
diferenca_meses = (
    periodo_conclusao.astype(int) - periodo_fim_previsto.astype(int)
)

# Classificação
df_restinga.loc[df_restinga["Situação de Matrícula"] == "Concluída", "Situação de Matrícula"] = np.where(
    diferenca_meses[df_restinga["Situação de Matrícula"] == "Concluída"] <= 5,
    "Concluída no prazo",
    "Concluída com atraso"
)

df_restinga["Situação de Matrícula"].value_counts()
```
